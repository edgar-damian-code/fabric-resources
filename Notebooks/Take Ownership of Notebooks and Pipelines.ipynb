```python
import sempy.fabric as fabric
import requests

def take_ownership_of_artifacts(workspace_id, service_account_token):
    headers = {
        'Authorization': f'Bearer {service_account_token}',
        'Content-Type': 'application/json'
    }

    # List all notebooks in the workspace
    notebooks = fabric.list_notebooks(workspace=workspace_id)
    for notebook in notebooks:
        notebook_id = notebook['id']
        url = f"https://api.fabric.microsoft.com/v1.0/workspaces/{workspace_id}/notebooks/{notebook_id}/takeover"
        response = requests.post(url, headers=headers)
        if response.status_code == 200:
            print(f"Ownership of notebook {notebook['name']} has been transferred.")
        else:
            print(f"Failed to transfer ownership of notebook {notebook['name']}.")

    # List all pipelines in the workspace
    pipelines = fabric.list_pipelines(workspace=workspace_id)
    for pipeline in pipelines:
        pipeline_id = pipeline['id']
        url = f"https://api.fabric.microsoft.com/v1.0/workspaces/{workspace_id}/pipelines/{pipeline_id}/takeover"
        response = requests.post(url, headers=headers)
        if response.status_code == 200:
            print(f"Ownership of pipeline {pipeline['name']} has been transferred.")
        else:
            print(f"Failed to transfer ownership of pipeline {pipeline['name']}.")

# Example usage
workspace_id = "your_workspace_id"
service_account_token = "your_service_account_token"

take_ownership_of_artifacts(workspace_id, service_account_token)
```
